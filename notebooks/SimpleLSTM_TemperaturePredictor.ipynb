```python
import pandas as pd

# Assuming the file is in the current working directory. If not, provide the full path.
try:
  df = pd.read_csv('hourly_ottawa_airpot_2021-2023.csv')
  print(df.head()) # Display the first few rows to verify
except FileNotFoundError:
  print("Error: File not found. Please ensure the file 'hourly_ottawa_airpot_2021-2023.csv' exists in the current directory or provide the correct path.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
temperatureList = df['Temp (°C)'].tolist()
temperatureList
```

```python
start = 0
end = 0+24
splitTemperatureList = []
while end < len(temperatureList):
    if not any(pd.isna(x) for x in temperatureList[start:end]) and not pd.isna(temperatureList[end]):
        splitTemperatureList.append((temperatureList[start:end], temperatureList[end]))

    start += 1
    end += 1

splitTemperatureList
```

```python
import torch

X = [i for i,_ in splitTemperatureList]
Y = [j for _,j in splitTemperatureList]

X = torch.FloatTensor(X)
Y = torch.FloatTensor(Y)

print("X shape:", X.shape)
print("Y shape:", Y.shape)
```

```python
# Check for NaN in X tensor
print("Number of NaN in X:", torch.isnan(X).sum().item())
print("Indices where NaN occurs in X:", torch.where(torch.isnan(X)))

# Check for NaN in Y tensor
print("Number of NaN in Y:", torch.isnan(Y).sum().item())
print("Indices where NaN occurs in Y:", torch.where(torch.isnan(Y)))
```

```python
def prepare_sequences(data, seq_length):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        if not any(pd.isna(x) for x in data[i:i+seq_length]) and not pd.isna(data[i+seq_length]):
            sequences.append(data[i:i+seq_length])
            targets.append(data[i+seq_length])
    return torch.tensor(sequences), torch.tensor(targets)

# Convert your temperature data into sequences
X, Y = prepare_sequences(temperatureList, 24)
print(X.shape)
print(Y.shape)
```

```python
import torch.nn as nn

class WeatherLSTM(nn.Module):
    def __init__(self, sequence_length=24, hidden_size=50):
        super(WeatherLSTM, self).__init__()
        self.sequence_length = sequence_length
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(
            input_size=1,    # Single "column" of features we're looking at. For each sequence of hours, we're only looking at temp
            hidden_size=hidden_size,  # Number of "brains" processing data. Each focuses on something and outputs a "signal" of what theyve detected
            batch_first=True # Expects input shape: (batch, seq_len, features)
        )
        self.linear = nn.Linear(hidden_size, 1)  # Take 50 signals from hidden layer and make final decision

    def forward(self, x):
        x = x.unsqueeze(-1)  # [[21,24...(24 temps here)], [(24 temps here)], ...] -> [[ [21], [22], [23], ... (24 temps in singleton lists)]]. At each hourly timestep, there's 1 feature to look at, temperature
        lstm_out, _ = self.lstm(x) # Get final output from hidden layer -> (26030, 24, 50)
        predictions = self.linear(lstm_out[:, -1, :]) # Dimension of lstm_out = (26030, 24, 50) -> for each 26030 sequence, look at the hidden output after the 24th hour, take each hidden signal
        # Then use linear layer to convert 50 signals into final decision into [26030, 1]
        return predictions

    def calculateError(self, predictions, targets):
        squeezedPredictions = predictions.squeeze(-1) # Predictions is [26030, 1], squeeze last dimension out
        return torch.mean((squeezedPredictions - targets)**2)
```

Using Mean Squared Error for temperature is a little strange because we're ignoring when a predicted temperature over or undershoots an actual value. Presumably you correct your network differently in those situations. Putting a pin in this for now.

```python
model = WeatherLSTM()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
def train_model_with_config(hidden_size, learning_rate, sequence_length):
    X, Y = prepare_sequences(temperatureList, sequence_length)

    # Split into train/val
    split_idx = int(len(X) * 0.8)
    X_train, X_val = X[:split_idx], X[split_idx:]
    Y_train, Y_val = Y[:split_idx], Y[split_idx:]

    model = WeatherLSTM(sequence_length, hidden_size)
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    train_losses = []
    val_losses = []

    for epoch in range(10):
        # Training
        model.train()
        optimizer.zero_grad()
        train_pred = model.forward(X_train)
        train_loss = model.calculateError(train_pred, Y_train)
        train_loss.backward()
        optimizer.step()

        # Validation (no gradients needed)
        model.eval()
        with torch.no_grad():
            val_pred = model.forward(X_val)
            val_loss = model.calculateError(val_pred, Y_val)

        # Store losses for plotting
        train_losses.append(math.sqrt(train_loss.item()))
        val_losses.append(math.sqrt(val_loss.item()))

        # print(f"Epoch {epoch}: Train RMSE = {train_losses[-1]:.2f}, Val RMSE = {val_losses[-1]:.2f}")

    return train_losses[-1], val_losses[-1]

# Test different sequence lengths with our best hyperparameters
configs = [
    (75, 0.02, 6),    # 6 hours of history
    (75, 0.02, 12),   # 12 hours
    (75, 0.02, 24),   # 24 hours (our original)
    (75, 0.02, 48),   # 2 days
    (75, 0.02, 72),   # 3 days
]

for hidden_size, lr, seq_len in configs:
    train_rmse, val_rmse = train_model_with_config(hidden_size, lr, seq_len)
    print(f"Seq Length: {seq_len}, Train RMSE: {train_rmse:.2f}, Val RMSE: {val_rmse:.2f}")
```

```python
def get_predictions(model, X_test, Y_test):
    model.eval()
    with torch.no_grad():
        predictions = model.forward(X_test)
    return predictions.squeeze(-1).numpy(), Y_test.numpy()

sequence_lengths = [6, 24, 72]
predictions = {}
actuals = {}
test_size = 100

for seq_len in sequence_lengths:
    print(f"\nTraining model with sequence length: {seq_len}")
    # Prepare data
    X, Y = prepare_sequences(temperatureList, seq_len)

    # Proper train/test split
    X_test = X[-test_size:]
    Y_test = Y[-test_size:]
    X_train = X[:-test_size]
    Y_train = Y[:-test_size]

    # Train model
    model = WeatherLSTM(sequence_length=seq_len, hidden_size=75)  # Fixed constructor args
    optimizer = torch.optim.Adam(model.parameters(), lr=0.02)

    for epoch in range(10):
        model.train()
        optimizer.zero_grad()
        train_pred = model.forward(X_train)
        train_loss = model.calculateError(train_pred, Y_train)
        train_loss.backward()
        optimizer.step()

    # Get predictions and actuals
    preds, acts = get_predictions(model, X_test, Y_test)
    predictions[f'{seq_len}h'] = preds
    actuals[f'{seq_len}h'] = acts

print("\nShapes for each sequence length:")
for seq_len in sequence_lengths:
    print(f"{seq_len}h: Predictions {predictions[f'{seq_len}h'].shape}, Actuals {actuals[f'{seq_len}h'].shape}")

print("\nFirst few predictions and actuals:")
for i in range(5):
    print(f"\nTimestep {i}:")
    for seq_len in sequence_lengths:
        print(f"{seq_len}h - Actual: {actuals[f'{seq_len}h'][i]:.2f}, Predicted: {predictions[f'{seq_len}h'][i]:.2f}")
```
